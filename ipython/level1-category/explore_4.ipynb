```python
import pandas as pd
```

```python
dis = pd.read_csv(
    "../../data/level2/m111-sku-dis-from-dw.csv", 
    sep=',', parse_dates=['order_date']
).rename(columns={'order_date': 'dis_date'})
```

```python
category = pd.read_csv("../../data/level2/item2category-from-3.csv", sep=',')
```

```python
dis = dis.join(category.set_index('item_code'), on='item_code', how='left')
```

```python
dis['month'] = dis.dis_date.astype('str').apply(lambda x: x[:7])
```

```python
dis.head()
```

```python
dis = dis.groupby(['category', 'month'])[['qty']].sum()
```

```python
dis = dis.unstack(level=-1).fillna(0)
```

```python
dis.columns = dis.columns.get_level_values(1)
```

```python
dis.to_csv("dis.txt", sep='\t')
```
